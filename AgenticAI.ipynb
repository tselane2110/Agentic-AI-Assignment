### Question 1 - Sigmoid activation with X1 = 16 and X2 = 11

````python
import numpy as np

# Activation functions - choose the required one
def sigmoid(x):
    return 1 / (1 + np.exp(-x))

# -------- INPUT --------
x1 = 16
x2 = 11

# -------- WEIGHTS --------

# Hidden Layer 1 weights
w_h1 = {
    "n1": {"x1": 0.1},
    "n2": {"x1": 0.3, "x2": 0.1},
    "n3": {"x1": 0.1, "x2": 0.1},
    "n4": {"x1": 0, "x2": 0.9},
}

b_h1 = {"n1": -1, "n2": -1, "n3": -1, "n4": -1}

# Hidden Layer 2 weights
w_h2 = {
    "n1": {"n2": 0.33, "n3": 0.2, "n4": 0.3},
    "n2": {"n1": 0.0},
    "n3": {"n2": 0.5, "n4": 0.33},
    "n4": {"n1": 0.33, "n3": 0.3},
}

b_h2 = {"n1": 1, "n2": 1, "n3": 1, "n4": 1}

# Output layer weights
w_out = {"n1": 0.1, "n2": 0.11, "n3": 0.4, "n4": 0.21}
b_out = 2

# -------- FORWARD PASS --------

# Hidden Layer 1
h1 = {}

h1["n1"] = sigmoid(x1 * w_h1["n1"]["x1"] + b_h1["n1"])
h1["n2"] = sigmoid(x1 * w_h1["n2"]["x1"] + x2 * w_h1["n2"]["x2"] + b_h1["n2"])
h1["n3"] = sigmoid(x1 * w_h1["n3"]["x1"] + x2 * w_h1["n3"]["x2"] + b_h1["n3"])
h1["n4"] = sigmoid(x1 * w_h1["n4"]["x1"] + x2 * w_h1["n4"]["x2"] + b_h1["n4"])

print("Hidden Layer 1:", h1)

# Hidden Layer 2
h2 = {}

h2["n1"] = sigmoid(
    h1["n2"] * w_h2["n1"]["n2"] +
    h1["n3"] * w_h2["n1"]["n3"] +
    h1["n4"] * w_h2["n1"]["n4"] +
    b_h2["n1"]
)

h2["n2"] = sigmoid(
    h1["n1"] * w_h2["n2"]["n1"] +
    b_h2["n2"]
)

h2["n3"] = sigmoid(
    h1["n2"] * w_h2["n3"]["n2"] +
    h1["n4"] * w_h2["n3"]["n4"] +
    b_h2["n3"]
)

h2["n4"] = sigmoid(
    h1["n1"] * w_h2["n4"]["n1"] +
    h1["n3"] * w_h2["n4"]["n3"] +
    b_h2["n4"]
)

print("Hidden Layer 2:", h2)

# Output Layer
output = sigmoid(
    h2["n1"] * w_out["n1"] +
    h2["n2"] * w_out["n2"] +
    h2["n3"] * w_out["n3"] +
    h2["n4"] * w_out["n4"] +
    b_out
)

print("Final Output:", output)
````
